```python
import pandas as pd
import cv2
import matplotlib.pyplot as plt
```

```python
def get_frame_count(video_path):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        # print(f"Error opening video file: {video_path}")
        return None
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    cap.release()
    return frame_count
    
```

```python
def train_video_filter(df):
    df = df.copy()
    # df = df[['START_REALIGNED','END_REALIGNED','SENTENCE']]
    # df['LENGTH'] = df['END_REALIGNED'] - df['START_REALIGNED']
    # df = df[['SENTENCE', 'LENGTH']]
    df['VIDEO_PATH'] = r"./train/" + df.index +'.mp4'
    df['JSON_PATH'] = r"./train/" + df.index +'/'
    
    df['FRAME_COUNT'] = df['VIDEO_PATH'].apply(get_frame_count)
    return df

def val_video_filter(df):
    df = df.copy()
    # df = df[['START_REALIGNED','END_REALIGNED','SENTENCE']]
    # df['LENGTH'] = df['END_REALIGNED'] - df['START_REALIGNED']
    # df = df[['SENTENCE', 'LENGTH']]
    df['VIDEO_PATH'] = r"./val/" + df.index +'.mp4'
    df['JSON_PATH'] = r"./val/" + df.index +'/'
    
    df['FRAME_COUNT'] = df['VIDEO_PATH'].apply(get_frame_count)
    return df

def test_video_filter(df):
    df = df.copy()
    # df = df[['START_REALIGNED','END_REALIGNED','SENTENCE']]
    # df['LENGTH'] = df['END_REALIGNED'] - df['START_REALIGNED']
    # df = df[['SENTENCE', 'LENGTH']]
    df['VIDEO_PATH'] = r"./test/" + df.index +'.mp4'
    df['JSON_PATH'] = r"./test/" + df.index +'/'
    df['FRAME_COUNT'] = df['VIDEO_PATH'].apply(get_frame_count)
    return df
```

```python
train_Data = pd.read_csv('./data33_train.csv', index_col='SENTENCE_NAME')
val_Data = pd.read_csv('./data33_val.csv', index_col='SENTENCE_NAME')
test_Data = pd.read_csv('./data33_test.csv', index_col='SENTENCE_NAME')
```

```python
train_Data = train_video_filter(train_Data)
train_Data = train_Data.dropna()

val_Data = val_video_filter(val_Data)
val_Data = val_Data.dropna()

test_Data = test_video_filter(test_Data)
test_Data = test_Data.dropna()
```

```python
def filter_main_df(train_Data):
# train_Data.sort_values('FRAME_COUNT')
    train_Data['FRAME_COUNT'].hist(bins=50)  # 可以调整 bins 的数量来改变直方图的粒度
    plt.title('Distribution of LENGTH')
    plt.xlabel('FRAME_COUNT')
    plt.ylabel('Frequency')
    plt.show()
    
    
    # 绘制 LENGTH 列的密度图
    train_Data['FRAME_COUNT'].plot(kind='density')
    plt.title('Density Plot of FRAME_COUNT')
    plt.xlabel('LENGFRAME_COUNTTH')
    plt.show()
    
    # 输出 LENGTH 列的主要分位数
    quantiles = train_Data['FRAME_COUNT'].quantile([0.25, 0.5, 0.75, 0.9, 0.95, 0.99])
    print(quantiles)
    
    # 计算90%分位数
    threshold = train_Data['FRAME_COUNT'].quantile(0.90)
    print("90%分位数阈值:", threshold)
    
    # 过滤掉超过阈值的数据
    filtered_Data = train_Data[train_Data['FRAME_COUNT'] <= threshold]
    
    print("过滤后的数据行数:", filtered_Data.shape[0])
    return filtered_Data
```
